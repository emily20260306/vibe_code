````python
# ==========================================
# 標題：Vibe Coding 檔案匯整 - 自動分類工具
# 功能：選取一個 Excel 檔案，依指定欄位自動拆分成多個工作表 (Sheet)
# ==========================================

import pandas as pd
from google.colab import files
from IPython.display import display, HTML
import io
import os

# 1. 顯示標題
display(HTML("<h1 style='color: #4285F4; font-family: sans-serif;'>Vibe Coding 檔案匯整 - 檔案分類</h1>"))

# 2. 上傳檔案
print("第一步：請上傳你想處理的 Excel 檔案 (.xlsx)")
uploaded = files.upload()

if not uploaded:
    print("❌ 未選取任何檔案，請重新執行。")
else:
    # 取得檔名
    filename = list(uploaded.keys())[0]
    # 讀取 Excel (讀取第一個工作表)
    df = pd.read_excel(io.BytesIO(uploaded[filename]))

    print(f"\n✅ 成功讀取檔案：{filename}")
    print("-" * 30)

    # 3. 讓使用者選取分類欄位
    columns = df.columns.tolist()
    print("第二步：請選擇你想用來「分類」的欄位（輸入數字編號）：")
    for i, col in enumerate(columns):
        print(f"[{i}] {col}")

    try:
        choice = int(input("\n請輸入欄位編號: "))
        selected_col = columns[choice]
        print(f"👉 你選擇了以「{selected_col}」進行分類")
    except Exception as e:
        print("❌ 輸入錯誤，請確保你輸入的是數字。")
        selected_col = None

    if selected_col:
        # 4. 開始處理分類
        output_filename = f"分類完成_{filename}"

        # 取得該欄位所有不重複的值
        unique_values = df[selected_col].dropna().unique()

        print(f"\n第三步：正在依據「{selected_col}」拆分資料...")

        # 使用 ExcelWriter 存檔
        with pd.ExcelWriter(output_filename, engine='openpyxl') as writer:
            for value in unique_values:
                # 篩選資料
                subset = df[df[selected_col] == value]

                # 清理工作表名稱 (不能超過 31 個字元，且不能有特殊符號)
                sheet_name = str(value)[:31].replace('[','').replace(']','').replace('*','').replace(':','').replace('?','').replace('/','').replace('\\','')

                # 寫入工作表
                subset.to_excel(writer, sheet_name=sheet_name, index=False)
                print(f"   - 已建立工作表: {sheet_name} (共 {len(subset)} 筆資料)")

        print("\n--- ✨ 處理完成！ ---")
        print(f"檔案已儲存為: {output_filename}")

        # 5. 下載檔案
        files.download(output_filename)

# --- 說明 ---
# 程式結束後，你會得到一個 Excel 檔案
# 裡面每個分頁 (Tab) 就是你所選欄位的一個分類結果
````
